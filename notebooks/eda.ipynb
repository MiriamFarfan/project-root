Limpieza de TotalCharges: convertir a númerico y remplazar espacíos vacios por NaN e imputar con la media 

```python
#Convertir a numerico
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df.info()
```

```python
print("Valores nulos en TotalCharges:", df["TotalCharges"].isnull().sum())
```

```python
#Imputar valores nulos con la mediana
mediana_total_charges = df["TotalCharges"].median()
print("Mediana:", mediana_total_charges)

#Imputa
df["TotalCharges"] = df["TotalCharges"].fillna(mediana_total_charges)

#Verifica
print("Nulos restantes:")
print(df["TotalCharges"].isna().sum())
```

Eliminar customerID, codificar gender, Partner, Churn a 0/1.

```python
#Eliminar variable que no aporta información relevante para el análisis- customerID
df = df.drop(columns=["customerID"])
print(df.columns)
```

Codificar variables binarias 

```python
#gender 
df["gender"] = df["gender"].map({"Male": 1, "Female": 0})

#Partner
df["Partner"] = df["Partner"].map({"Yes": 1, "No": 0})

#Churn
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
print(df.head())
```

```python
print(df[["gender", "Partner", "Churn"]].isna().sum())
```

```python
#Verificación 
print(df[["gender", "Partner", "Churn"]].head())
print(df[["gender", "Partner", "Churn"]].dtypes)
```

```python
#Analsis de otras variables categoricas 
categorical_columns = df.select_dtypes(include=["object"]).columns
print(categorical_columns)
```

```python
for col in categorical_columns:
    print(f"\n===== {col} =====")
    print(df[col].unique())
```

```python
binary_columns = ["Dependents","PhoneService","PaperlessBilling"]
for col in binary_columns:
    df[col] = df[col].map({"Yes": 1, "No": 0})
print(df[binary_columns].head())
```

```python
#Columnas multiclase 
multiclass_columns =["MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup","DeviceProtection", "TechSupport","StreamingTV", "StreamingMovies", "Contract", "PaymentMethod"]
print(multiclass_columns)
```

```python
#Prueba de codificación One-Hot Encoding
df = pd.get_dummies(df, columns=multiclass_columns, drop_first=False)
print(df.head())
```

```python
print("Nulos totales en el dataframe:", df.isna().sum().sum())
```
